# 여러가지의 그래프를 가지는 경우
- muti-variable이라고 함
- Hypothesis = w1x1 + w2x2 + w3x3 + wnxn + b
- Hypothesis는 matrix를 활용하여 구함
- x는 매트릭스(2차원)를 구성하고 w는 매트릭스의 열의 개수에 맞는 행을 구성
- x(a, b)라면 w(b, 1)를 구성  --> a를 인스턴스라고 지칭 // b를 매트릭스라 지칭
- X(5, 3) * W(3, 1) = H(X)(5, 1) 구성 // H(X)의 행은 x의 행과 일치
- -> 우리가 설계하는 부분은 W이므로 그 부분을 설계하는데 X의 열과 H(X)의 열의 갯수에 맞추면됨
- -> W의 행은 한개의 데이터 안에 있는 tensor 값을 지칭 // 열은 데이터가 들어가 최종 출력받고자하는 output 값을 지칭

# muti-variable 예제(매트릭스 변환 전)

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]
```

```python
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)
```

```python
w1 = tf.Variable(tf.random_normal([1]), name ="weight1")
w2 = tf.Variable(tf.random_normal([1]), name ="weight2")
w3 = tf.Variable(tf.random_normal([1]), name ="weight3")
b = tf.Variable(tf.random_normal([1]), name ="bias")
```

```python
hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPredction:\n", hy_val)
```

# muti-variable 예제(매트릭스 변환 후)

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

## 매트릭스 변환
- x_data 3개의 값(5열)을 5행 3열로 변환
- y_data 값(5열)을 5행으로 변환

## x_data 구성 측면에서 살펴보기
- 5행 인스턴스(data개수) -  5개
- 3열 data 내부 구조 - shape를 지칭 shape=(None,3) // None으로 표기 이유 -> vector(1차원)이므로  만약 matrix(2차원)라면 shape=[1,3] 

```python
x_data = [[73., 80., 75.], [93., 88., 93.], [89., 91., 90.], [96.,98., 100.], [73. , 66., 70.]]
y_data = [[152.], [185.], [180.], [196.], [142.]]
```

## placeholder -> data를 맵핑하는 역할

## shape의 구성
- X(data)의 경우 shape는 data하나가 구성하고 있는 shape를 사용
- Y(output / 결과 값)의 경우 shape는 얻고자하는 정답 data의 갯수를 의미(정답 1개일 땐 1개 // 다수일 땐 다수개) 

```python
X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])
```

## Variable -> data를 tensor로 변환

## W(가중치)의 shape만드는 방법
- 각각의 shape 값을 곱해주면됨 
- 예를 들어: data(3,5)라면 W는 열과 행을 반대로(5,3)로 만들어주어 가중치 적용

###### muti-variable를 활용 할 때는 X*W를 하는 것이 아니라 matmul을 활용하여 값이 적절하게 곱해질 수 있게 만들어줌(broadcasting과 차이 파악)

```python
W = tf.Variable(tf.random_normal([3, 1]), name ="weight1")
b = tf.Variable(tf.random_normal([1]), name ="bias")
```

```python
hypothesis = tf.matmul(X, W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={X: x_data, Y: y_data})
    if step % 10 == 0:
        print(step, "Cost: ", cost_val, "\nPredction:\n", hy_val)
```
